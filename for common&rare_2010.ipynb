```python
# Define rainy days (1 if PRCP_mm > 0, else 0)
df['is_rainy'] = (df['PRCP_mm'] > 0 ).astype(int)

# Check
df[['date', 'PRCP_mm', 'is_rainy']].head(10)
```

```python
# Define dry days (0 if PRCP_mm == 0, else 1)
df['is_dry'] = (df['PRCP_mm'] == 0 ).astype(int)

# Check
df[['date', 'PRCP_mm', 'is_dry']].head(10)
```

```python
# Focus on a single year
df_2010 = df[df['year'] == 2010].copy()

# Define hot days: MAX_C > 35
df_2010['is_hot'] = (df_2010['MAX_C'] > 35).astype(int)

# Joint event: hot AND dry
df_2010['hot_and_dry'] = ((df_2010['is_hot'] == 1) & (df_2010['is_dry'] == 1)).astype(int)

# Calculate joint probability
p_hot_and_dry = df_2010['hot_and_dry'].mean()

print(f"Probability of a hot and dry day in 2010: {p_hot_and_dry:.4f}")
```

```python
# Focus on a single year
df_2010 = df[df['year'] == 2010].copy()

# Define cool days: MIN_C < 25
df_2010['is_cool'] = (df_2010['MIN_C'] < 25).astype(int)

# Joint event: cool AND dry
df_2010['cool_and_dry'] = ((df_2010['is_cool'] == 1) & (df_2010['is_dry'] == 1)).astype(int)

# Calculate joint probability
p_cool_and_dry = df_2010['cool_and_dry'].mean()

print(f"Probability of a cool and dry day in 2010: {p_cool_and_dry:.4f}")
```
