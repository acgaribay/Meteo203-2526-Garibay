```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

# Robust and Resistant Methods

```python
# Part 1: Robust Methods
# Generate a simple dataset with outliers
np.random.seed(42)
data = np.random.normal(50, 10, 100)  # Normally distributed data
data_with_outliers = np.append(data, [150, 160, 170])  # Add outliers

# Calculate mean and median
mean_original = np.mean(data)
median_original = np.median(data)

print(f"Original Mean: {mean_original}, Original Median: {median_original}")
```

```python
mean_with_outliers = np.mean(data_with_outliers)
median_with_outliers = np.median(data_with_outliers)

print(f"Mean with Outliers: {mean_with_outliers}, Median with Outliers: {median_with_outliers}")
```

```python
# Part 2: Resistant Measures
# Calculate standard deviation and IQR (interquartile range)
std_original = np.std(data)
iqr_original = np.percentile(data, 75) - np.percentile(data, 25)

print(f"Original Standard Deviation: {std_original}, Original IQR: {iqr_original}")


```

```python
std_with_outliers = np.std(data_with_outliers)
iqr_with_outliers = np.percentile(data_with_outliers, 75) - np.percentile(data_with_outliers, 25)

print(f"Standard Deviation with Outliers: {std_with_outliers}, IQR with Outliers: {iqr_with_outliers}")
```

```python
# Part 3: Visual Demonstrations
# Plot histograms for original data and data with outliers
plt.figure(figsize=(14, 6))

# Histogram for original data
plt.subplot(1, 2, 1)
sns.histplot(data, bins=20, color='blue', kde=True)
plt.title("Histogram of Original Data (No Outliers)")
plt.axvline(mean_original, color='r', linestyle='--', label=f'Mean: {mean_original:.2f}')
plt.axvline(median_original, color='g', linestyle='--', label=f'Median: {median_original:.2f}')
plt.legend()

# Histogram for data with outliers
plt.subplot(1, 2, 2)
sns.histplot(data_with_outliers, bins=20, color='orange', kde=True)
plt.title("Histogram of Data with Outliers")
plt.axvline(mean_with_outliers, color='r', linestyle='--', label=f'Mean: {mean_with_outliers:.2f}')
plt.axvline(median_with_outliers, color='g', linestyle='--', label=f'Median: {median_with_outliers:.2f}')
plt.legend()

plt.show()
```

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Part 1: Generate normally distributed data (no skew)
data_normal = np.random.normal(50, 10, 1000)

# Part 2: Generate data with skew (by adding extreme values)
data_skewed = np.append(data_normal, [200, 220, 250, 300, 350])    # Adding outliers to create skewness

# Calculate mean and median for both datasets
mean_normal = np.mean(data_normal)
median_normal = np.median(data_normal)
mean_skewed = np.mean(data_skewed)
median_skewed = np.median(data_skewed)

# Part 3: Plot histograms for normal and skewed data, with mean and median
plt.figure(figsize=(14, 6))

# Histogram for normally distributed data (no skew)
plt.subplot(1, 2, 1)
sns.histplot(data_normal, bins=30, color='skyblue', kde=True)
plt.title('Histogram of Normally Distributed Data')
plt.axvline(mean_normal, color='red', linestyle='--', label=f'Mean: {mean_normal:.2f}')
plt.axvline(median_normal, color='green', linestyle='--', label=f'Median: {median_normal:.2f}')
plt.legend()

# Histogram for skewed data (with outliers)
plt.subplot(1, 2, 2)
sns.histplot(data_skewed, bins=30, color='lightcoral', kde=True)
plt.title('Histogram of Skewed Data (with Outliers)')
plt.axvline(mean_skewed, color='red', linestyle='--', label=f'Mean: {mean_skewed:.2f}')
plt.axvline(median_skewed, color='green', linestyle='--', label=f'Median: {median_skewed:.2f}')
plt.legend()

plt.tight_layout()
plt.show()

```

# Median, Trimean, Trimmed Mean

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generate random datasets
np.random.seed(42)
data_normal = np.random.normal(50, 10, 1000)  # Normally distributed data
data_skewed = np.append(np.random.normal(50, 10, 950), [100, 110, 120, 130, 150])  # Skewed data with moderate outliers
data_extreme = np.append(np.random.normal(50, 10, 900), [200, 250, 300, 350, 400])  # Data with extreme outliers

# Calculate median, trimean, and trimmed mean
median_normal = np.median(data_normal)
trimean_skewed = (np.percentile(data_skewed, 25) + 2 * np.median(data_skewed) + np.percentile(data_skewed, 75)) / 4
trimmed_mean_extreme = stats.trim_mean(data_extreme, 0.1)  # Trim 10% from both ends

# Create subplots to demonstrate where median, trimean, and trimmed mean are useful
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Median - useful in normally distributed data
sns.histplot(data_normal, bins=30, color='skyblue', kde=True, ax=axs[0])
axs[0].axvline(median_normal, color='green', linestyle='--', label=f'Median: {median_normal:.2f}')
axs[0].set_title('Normally Distributed Data (Median)')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')
axs[0].legend()

# Plot 2: Trimean - useful in skewed data with moderate outliers
sns.histplot(data_skewed, bins=30, color='lightcoral', kde=True, ax=axs[1])
axs[1].axvline(trimean_skewed, color='blue', linestyle='--', label=f'Trimean: {trimean_skewed:.2f}')
axs[1].set_title('Skewed Data with Moderate Outliers (Trimean)')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')
axs[1].legend()

# Plot 3: Trimmed Mean - useful in data with extreme outliers
sns.histplot(data_extreme, bins=30, color='lightgreen', kde=True, ax=axs[2])
axs[2].axvline(trimmed_mean_extreme, color='purple', linestyle='--', label=f'Trimmed Mean: {trimmed_mean_extreme:.2f}')
axs[2].set_title('Data with Extreme Outliers (Trimmed Mean)')
axs[2].set_xlabel('Value')
axs[2].set_ylabel('Frequency')
axs[2].legend()

plt.tight_layout()
plt.show()
```

# Interquartile Range vs Standard Deviation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random datasets
np.random.seed(42)
data_normal = np.random.normal(50, 10, 1000)  # Normally distributed data
data_with_outliers = np.append(np.random.normal(50, 10, 950), [100, 110, 120, 130, 150])  # Skewed data with outliers

# Calculate IQR and standard deviation for both datasets
iqr_normal = np.percentile(data_normal, 75) - np.percentile(data_normal, 25)
std_normal = np.std(data_normal)

iqr_outliers = np.percentile(data_with_outliers, 75) - np.percentile(data_with_outliers, 25)
std_outliers = np.std(data_with_outliers)

# Create subplots to compare IQR and Standard Deviation
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Normal distribution
sns.histplot(data_normal, bins=30, kde=True, color='skyblue', ax=axs[0])
axs[0].axvline(np.mean(data_normal), color='red', linestyle='--', label=f'Mean: {np.mean(data_normal):.2f}')
axs[0].axvline(np.median(data_normal), color='green', linestyle='--', label=f'Median: {np.median(data_normal):.2f}')
axs[0].set_title('Normally Distributed Data')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')
axs[0].legend()

# Add IQR and standard deviation info
axs[0].text(55, 50, f'IQR: {iqr_normal:.2f}\nStd Dev: {std_normal:.2f}', color='black', fontsize=12)

# Plot 2: Skewed data with outliers
sns.histplot(data_with_outliers, bins=30, kde=True, color='lightcoral', ax=axs[1])
axs[1].axvline(np.mean(data_with_outliers), color='red', linestyle='--', label=f'Mean: {np.mean(data_with_outliers):.2f}')
axs[1].axvline(np.median(data_with_outliers), color='green', linestyle='--', label=f'Median: {np.median(data_with_outliers):.2f}')
axs[1].set_title('Skewed Data with Outliers')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')
axs[1].legend()

# Add IQR and standard deviation info
axs[1].text(55, 50, f'IQR: {iqr_outliers:.2f}\nStd Dev: {std_outliers:.2f}', color='black', fontsize=12)

plt.tight_layout()
plt.show()
```

# MAD and Trimmed Variance

```python
# Load the dataset
file_path = '../../Meteo203-2526-Garibay/data/2009_SCIENCE_GARDEN.csv'
df = pd.read_csv(file_path)

# Clean up column names if needed and drop missing values in 'DAILY RAINFALL'
df.columns = df.columns.str.strip()  # Remove any extra whitespace
rainfall_data = df['DAILY RAINFALL'].dropna()

# Exclude non-rainy days (remove zero values)
rainfall_data_nonzero = rainfall_data[rainfall_data > 0]

# Part 1: Calculate stats for all days (with zeros)
median_rainfall = np.median(rainfall_data)
iqr_rainfall = np.percentile(rainfall_data, 75) - np.percentile(rainfall_data, 25)
mad_rainfall = stats.median_abs_deviation(rainfall_data)
trimmed_variance_rainfall = stats.tvar(rainfall_data, limits=(0.1, 0.9))  # Trim 10% from both ends
# Part 2: Calculate stats for rainy days only (non-zero values)
median_rainfall_nonzero = np.median(rainfall_data_nonzero)
iqr_rainfall_nonzero = np.percentile(rainfall_data_nonzero, 75) - np.percentile(rainfall_data_nonzero, 25)
mad_rainfall_nonzero = stats.median_abs_deviation(rainfall_data_nonzero)
trimmed_variance_rainfall_nonzero = stats.tvar(rainfall_data_nonzero, limits=(0.1, 0.9))  # Trim 10% from both ends


# Display the results for all days
print("All Days (Including Non-Rainy):")
print(f'Median: {median_rainfall}')
print(f'IQR: {iqr_rainfall}')
print(f'MAD: {mad_rainfall}')
print(f'Trimmed Variance: {trimmed_variance_rainfall}')

# Display the results for rainy days only
print("\nRainy Days Only (Zeroes Excluded):")
print(f'Median (Non-Zero): {median_rainfall_nonzero}')
print(f'IQR (Non-Zero): {iqr_rainfall_nonzero}')
print(f'MAD (Non-Zero): {mad_rainfall_nonzero}')
print(f'Trimmed Variance (Non-Zero): {trimmed_variance_rainfall_nonzero}')

# Part 3: Visualize the rainfall distribution (all days)
plt.figure(figsize=(10, 6))
sns.histplot(rainfall_data, bins=30, kde=True, color='blue')
plt.axvline(median_rainfall, color='green', linestyle='--', label=f'Median: {median_rainfall:.2f}')
plt.title('Rainfall Distribution (All Days, Including Zeroes)')
plt.xlabel('Daily Rainfall (mm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Part 4: Visualize the rainfall distribution (rainy days only)
plt.figure(figsize=(10, 6))
sns.histplot(rainfall_data_nonzero, bins=30, kde=True, color='orange')
plt.axvline(median_rainfall_nonzero, color='green', linestyle='--', label=f'Median: {median_rainfall_nonzero:.2f}')
plt.title('Rainfall Distribution (Rainy Days Only, Zeroes Excluded)')
plt.xlabel('Daily Rainfall (mm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Symmetry: skewness coefficient vs Yule-Kendall Index




```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic datasets with varying skewness
np.random.seed(42)
data_symmetric = np.random.normal(50, 10, 1000)  # Symmetric data (normally distributed)
data_positively_skewed = np.random.exponential(10, 1000) + 40  # Positively skewed data
data_negatively_skewed = 100 - np.random.exponential(10, 1000)  # Negatively skewed data

# Function to calculate Yule-Kendall Index (YK)
def yule_kendall_index(data):
    q1 = np.percentile(data, 25)
    median = np.median(data)
    q3 = np.percentile(data, 75)
    yk_index = (q3 + q1 - 2 * median) / (q3 - q1)
    return yk_index

# Calculate skewness and YK index for each dataset
datasets = {
    'Symmetric': data_symmetric,
    'Positively Skewed': data_positively_skewed,
    'Negatively Skewed': data_negatively_skewed
}

for name, data in datasets.items():
    skewness = stats.skew(data)
    yk_index = yule_kendall_index(data)
    print(f'{name} Data:')
    print(f'  Skewness Coefficient: {skewness:.3f}')
    print(f'  Yule-Kendall Index: {yk_index:.3f}')
    print('-' * 30)

# Plot histograms for visual comparison
plt.figure(figsize=(15, 5))

# Plot 1: Symmetric Data
plt.subplot(1, 3, 1)
sns.histplot(data_symmetric, bins=30, kde=True, color='blue')
plt.title('Symmetric Data')

# Plot 2: Positively Skewed Data
plt.subplot(1, 3, 2)
sns.histplot(data_positively_skewed, bins=30, kde=True, color='green')
plt.title('Positively Skewed Data')

# Plot 3: Negatively Skewed Data
plt.subplot(1, 3, 3)
sns.histplot(data_negatively_skewed, bins=30, kde=True, color='red')
plt.title('Negatively Skewed Data')

plt.tight_layout()
plt.show()
```

# Cumulative Frequency Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic datasets with varying skewness
np.random.seed(42)
data_symmetric = np.random.normal(50, 10, 1000)  # Symmetric data (normally distributed)
data_positively_skewed = np.random.exponential(10, 1000) + 40  # Positively skewed data
data_negatively_skewed = 100 - np.random.exponential(10, 1000)  # Negatively skewed data

# Sort the data for CDF plotting
data_symmetric_sorted = np.sort(data_symmetric)
data_positively_skewed_sorted = np.sort(data_positively_skewed)
data_negatively_skewed_sorted = np.sort(data_negatively_skewed)

# Generate CDF values for each dataset
cdf_symmetric = np.arange(1, len(data_symmetric_sorted) + 1) / len(data_symmetric_sorted)
cdf_positively_skewed = np.arange(1, len(data_positively_skewed_sorted) + 1) / len(data_positively_skewed_sorted)
cdf_negatively_skewed = np.arange(1, len(data_negatively_skewed_sorted) + 1) / len(data_negatively_skewed_sorted)

# Create the histograms and CDF plots
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Plot 1: Symmetric Data - Histogram
sns.histplot(data_symmetric, bins=30, kde=True, color='blue', ax=axs[0, 0])
axs[0, 0].set_title('Histogram - Symmetric Data')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')

# Plot 2: Symmetric Data - CDF
axs[0, 1].plot(data_symmetric_sorted, cdf_symmetric, label='CDF', color='blue')
axs[0, 1].set_title('CDF - Symmetric Data (S-shaped)')
axs[0, 1].set_xlabel('Value')
axs[0, 1].set_ylabel('Cumulative Probability')

# Plot 3: Positively Skewed Data - Histogram
sns.histplot(data_positively_skewed, bins=30, kde=True, color='green', ax=axs[1, 0])
axs[1, 0].set_title('Histogram - Positively Skewed Data')
axs[1, 0].set_xlabel('Value')
axs[1, 0].set_ylabel('Frequency')

# Plot 4: Positively Skewed Data - CDF
axs[1, 1].plot(data_positively_skewed_sorted, cdf_positively_skewed, label='CDF', color='green')
axs[1, 1].set_title('CDF - Positively Skewed Data')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Cumulative Probability')

# Plot 5: Negatively Skewed Data - Histogram
sns.histplot(data_negatively_skewed, bins=30, kde=True, color='red', ax=axs[2, 0])
axs[2, 0].set_title('Histogram - Negatively Skewed Data')
axs[2, 0].set_xlabel('Value')
axs[2, 0].set_ylabel('Frequency')

# Plot 6: Negatively Skewed Data - CDF
axs[2, 1].plot(data_negatively_skewed_sorted, cdf_negatively_skewed, label='CDF', color='red')
axs[2, 1].set_title('CDF - Negatively Skewed Data')
axs[2, 1].set_xlabel('Value')
axs[2, 1].set_ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```
