# Power Transformation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate skewed data (e.g., exponential distribution)
np.random.seed(42)
skewed_data = np.random.exponential(2, 1000) + 1  # Highly skewed data

# Apply log transformation (log reexpression)
log_transformed_data = np.log(skewed_data)

# Plotting the original skewed data and log-transformed data side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Original Skewed Data
sns.histplot(skewed_data, bins=30, kde=True, color='red', ax=axs[0])
axs[0].set_title('Original Skewed Data')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

# Plot 2: Log Transformed Data
sns.histplot(log_transformed_data, bins=30, kde=True, color='blue', ax=axs[1])
axs[1].set_title('Log-Transformed Data (Reexpression)')
axs[1].set_xlabel('Log(Value)')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Standardized Anomalies

Standardized anomalies help reexpress the data by providing information about the **magnitude** of the **anomaly**, scaled according to the **standard deviation**.

In climatology,it can be expressed as $\frac{X-\mu}{\sigma}$ where 
- $X$ is the sampe point
- $\mu$ is the climatological mean
- $\sigma$ is the climatological standard deviation.

For more information: 
- https://actuariesclimateindex.org/wp-content/uploads/2016/04/SampleCalcEng.5.18.pdf
- https://iridl.ldeo.columbia.edu/dochelp/StatTutorial/Climatologies/index.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/2009_SCIENCE_GARDEN.csv'
df = pd.read_csv(file_path)

# Clean up column names if needed and drop missing values in 'TEMP MEAN'
df.columns = df.columns.str.strip()  # Remove any extra whitespace
temperature_data = df['TEMP MEAN'].dropna()  # Get the temperature data

# Convert 'DATE' column to datetime format if not already
df['DATE'] = pd.to_datetime(df['DATE'])

# Calculate mean and standard deviation of the temperature
mean_temp = np.mean(temperature_data)
std_temp = np.std(temperature_data)

# Calculate standardized anomalies (Z-scores)
df['TEMP_MEAN_Anomaly'] = (temperature_data - mean_temp) / std_temp

# Plot the original temperature data and the standardized anomalies
fig, axs = plt.subplots(2, 1, figsize=(12, 8))

# Plot 1: Original Mean Temperature Data
axs[0].plot(df['DATE'], df['TEMP MEAN'], label='Mean Temperature', color='blue')
axs[0].set_title('Mean Temperature Over Time (Science Garden)')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Temperature (°C)')
axs[0].legend()

# Plot 2: Standardized Anomalies (Z-scores)
axs[1].plot(df['DATE'], df['TEMP_MEAN_Anomaly'], label='Standardized Anomalies (Z-Scores)', color='orange')
axs[1].axhline(0, color='green', linestyle='--', label='Z = 0')  # Add the line at z=0
axs[1].set_title('Standardized Anomalies of Mean Temperature')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Temperature Anomaly (Z-Score)')
axs[1].legend()

plt.tight_layout()
plt.show()
```

# Absolute Anomalies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/2009_SCIENCE_GARDEN.csv'
df = pd.read_csv(file_path)

# Clean up column names if needed and drop missing values in 'TEMP MEAN'
df.columns = df.columns.str.strip()  # Remove any extra whitespace
temperature_data = df['TEMP MEAN'].dropna()  # Get the temperature data

# Convert 'DATE' column to datetime format if not already
df['DATE'] = pd.to_datetime(df['DATE'])

# Calculate mean and standard deviation of the temperature
mean_temp = np.mean(temperature_data)
std_temp = np.std(temperature_data)

# Calculate absolute anomalies (difference from the mean)
df['TEMP_MEAN_Absolute_Anomaly'] = temperature_data - mean_temp

# Calculate standardized anomalies (Z-scores)
df['TEMP_MEAN_Standardized_Anomaly'] = (temperature_data - mean_temp) / std_temp

# Plot the absolute and standardized anomalies side by side
fig, axs = plt.subplots(2, 1, figsize=(12, 8))

# Plot 1: Absolute Temperature Anomalies
axs[0].plot(df['DATE'], df['TEMP_MEAN_Absolute_Anomaly'], label='Absolute Anomalies', color='blue')
axs[0].axhline(0, color='green', linestyle='--', label='Mean Temperature')  # Add line at the mean
axs[0].set_title('Absolute Anomalies of Mean Temperature (°C)')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Temperature Anomaly (°C)')
axs[0].legend()

# Plot 2: Standardized Temperature Anomalies (Z-Scores)
axs[1].plot(df['DATE'], df['TEMP_MEAN_Standardized_Anomaly'], label='Standardized Anomalies (Z-Scores)', color='orange')
axs[1].axhline(0, color='green', linestyle='--', label='Z = 0')  # Add line at z=0
axs[1].set_title('Standardized Anomalies of Mean Temperature (Z-Scores)')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Z-Score')
axs[1].legend()

plt.tight_layout()
plt.show()
```
